```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "http://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
# 3. download html with a get request
response = requests.get(url)
```

```python
response.status_code # 200 status code means OK!
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup.select("body > div.container > div:nth-child(4) > div.col-xs-12.col-md-8 > div.chart-wrapper > li")
```

```python
for p in soup.select("p.title-artist"):
    print(p.get_text())
```

```python
print(soup.prettify())
```

```python
# You can find this code  <div class="chart-wrapper">

# The CSS selector div.chart-wrapper selects the <div> element with the class "chart-wrapper."

# In general, CSS selectors are used to style or manipulate specific elements on a web page. 
# The selector div.chart-wrapper specifically targets div elements with the class "chart-wrapper." The styling or manipulation rules associated with this selector would apply only to those elements that match the criteria.
```

```python
soup.select("body > div.container > div:nth-child(4) > div.col-xs-12.col-md-8 > div.chart-wrapper")
```

```python
# Search for the title and the artist
```

```python
artist=[]
for art in soup.select('.artist'):
    artist.append(art.get_text())
```

```python
artist
```

```python
title=[]
for tit in soup.select('.title'):
    title.append(tit.get_text())
```

```python
title
```

```python
# Put them into one dataframe
title_artist_top_100_df = pd.DataFrame({"title":title,
                          "artist":artist})
```

```python
title_artist_top_100_df
```

```python
# https://www.geeksforgeeks.org/randomly-select-columns-from-pandas-dataframe/
# # randomly selecting columns 
# df = df.sample(axis='columns')

# if song_title in concatenated_df['title'].values:
    # checks if the specified song_title exists in the 'title' column of the DataFrame concatenated_df. 
    # The "in" operator checks for membership in the list of titles
    
# If the condition the song title is found, this line returns a random song title from the 'title' column. 
    # concatenated_df['title']: This extracts the 'title' column from the DataFrame.
    # .sample(1) randomly samples one value from the 'title' column
    # .values[0]: This extracts the actual value from the resulting Series


def check_song_title(song_title):
    if song_title in concatenated_df['title'].values:
        return concatenated_df['title'].sample(1).values[0]
    else:
        return 'Sorry I can not make a recommondation at the moment'
```

in Operator:

The in operator checks for membership. In this case, it checks if song_title is present in the list of titles (concatenated_df['title'].values).
sample Method:

The sample method is used to randomly sample values from a Series (in this case, the 'title' column). The argument 1 specifies that we want to sample one value.
.values:

The .values attribute is used to get the underlying numpy array of a pandas Series. In this context, it's used to get the random sampled value as a numpy array.
Indexing [0]:

The [0] indexing is used to extract the first (and only) element from the numpy array obtained after sampling. This is necessary because sample(1) returns a pandas Series with one element, and we use [0] to get the actual value.

```python
check_song_title('White Horse')
```
