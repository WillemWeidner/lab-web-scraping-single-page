```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "http://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
# 3. download html with a get request
response = requests.get(url)
```

```python
response.status_code # 200 status code means OK!
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup.select("body > div.container > div:nth-child(4) > div.col-xs-12.col-md-8 > div.chart-wrapper > li")
```

```python
for p in soup.select("p.title-artist"):
    print(p.get_text())
```

```python
print(soup.prettify())
```

```python
# You can find this code  <div class="chart-wrapper">

# The CSS selector div.chart-wrapper selects the <div> element with the class "chart-wrapper."

# In general, CSS selectors are used to style or manipulate specific elements on a web page. 
# The selector div.chart-wrapper specifically targets div elements with the class "chart-wrapper." The styling or manipulation rules associated with this selector would apply only to those elements that match the criteria.
```

```python
soup.select("body > div.container > div:nth-child(4) > div.col-xs-12.col-md-8 > div.chart-wrapper")
```

```python
# Search for the title and the artist
```

```python
artist=[]
for art in soup.select('.artist'):
    artist.append(art.get_text())
```

```python
artist
```

```python
title=[]
for tit in soup.select('.title'):
    title.append(tit.get_text())
```

```python
title
```

```python
# Put them into one dataframe
title_artist_top_100_df = pd.DataFrame({"title":title,
                          "artist":artist})
```

```python
title_artist_top_100_df
```
